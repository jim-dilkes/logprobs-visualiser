```python
import sys
sys.path.append("..")

import json, os
from render_logprobs import render_logprobs

filedir = 'examples_data'

def load_render_logprobs(filename, dir):
    filename = filename + ".json" if not filename.endswith(".json") else filename
    load_path = os.path.join(dir, filename)

    with open(load_path, 'r') as f:
        examples = json.load(f)

    for ex in examples:
        render_logprobs(ex['logprobs'], ex['question'])
```
